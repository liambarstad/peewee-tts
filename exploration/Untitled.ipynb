```python
import re
from os import listdir
from os.path import isfile, join
import warnings
warnings.filterwarnings('ignore')

import moviepy.editor as mp

def convert_vid_file(vid_path, dest_path):
    mp_audio = mp.AudioFileClip(vid_path)
    mp_audio.write_audiofile(f'{dest_path}') 

files = [f for f in listdir('../data') if isfile(join('../data', f))]
for f in files:
    match = re.findall(r'S\d{2}E\d{2}', f)
    if len(match) > 0:
        season = match[0][1:3]
        episode = match[0][-2:]
        convert_vid_file(f'../data/{f}', f'../data/pee_wees_playhouse_{season}_{episode}.wav')
```

```python
convert_vid_file("../data/Pee-wee's.Big.Adventure.1985.1080p.BluRay.x264-[YTS.AG].mp4", '../data/movies/pee_wees_big_adventure.wav')



```
